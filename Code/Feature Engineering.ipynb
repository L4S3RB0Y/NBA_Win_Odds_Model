## Feature Engineering

1. Pre-Game std plus minus stats
- Compute the standard deviation of the +/- statistic over previous games.
- Higher variability in +/- indicates inconsistency, while lower variability suggests stability.
2. Home vs. Away std plus minus stats (Optional)
- Separate the +/- statistic into home and away games to capture the impact of home advantage.
- Compute averages and standard deviations for home and away games separately.
3. Home Game
- If the game is the home game, 1; otherwise 0.
4. Previous Total Win
- Total Number of win before match against the specific opponent.
5. Weight
- A recent game has a higher weight, which is the number of game starting from oldest to latest game. If PHX has played 10 games, the latest game has a weight 10.
6. FG_Efficiency
- efficieny of free throws made over all Free throws attempts

## stats stablibity

The standard deviation of stats based on previous games

```python
# Previous Game Total Wins

# extract the opponent out
def extract_opponent(row):
    matchup = row['Match Up']
    team = row['Team']
    if "vs." in matchup:
        return matchup.split("vs.")[1].strip()
    elif "@" in matchup:
        return matchup.split("@")[1].strip()
    return None

df['Opponent'] = df.apply(extract_opponent, axis=1)
```

```python
colnames = df.columns
colnames = colnames.drop(["Team", "Match Up", "Game Date", "W/L", "home_game", "Opponent"])
```

```python
colnames
```

```python
def pre_average_calc_with_opponent(colnames):
  team = df.Team.unique()

  for colname in colnames:
    print(colname)
    prev_mean_value = []
    opponent_prev_mean_value = []

    for team_name in team:
      subset = df[df['Team'] == team_name].reset_index(drop=True)

      for idx, row in subset.iterrows():
        if idx == 0:
          prev_mean_value.append(0)
          opponent_prev_mean_value.append(0)  # Initialize opponent's pre-mean
        else:
          prev_row = subset.loc[:idx - 1]
          prev_mean = np.mean(prev_row.loc[:, colname])
          prev_mean_value.append(prev_mean)

          # Get opponent's pre-mean
          opponent = row['Opponent']
          opponent_prev_games = df[(df['Team'] == opponent) & (df['Game Date'] < row['Game Date'])]
          if len(opponent_prev_games) > 0:
            opponent_prev_mean = np.mean(opponent_prev_games[colname])
            opponent_prev_mean_value.append(opponent_prev_mean)
          else:
            opponent_prev_mean_value.append(0)  # Handle cases where opponent has no previous games

    name = "Pre_mean_" + colname
    df[name] = prev_mean_value
    opponent_name = "Opponent_Pre_mean_" + colname  # New column for opponent's pre-mean
    df[opponent_name] = opponent_prev_mean_value
    pre_difference = "Pre_mean_diff_" + colname
    df[pre_difference] = df[name] - df[opponent_name]

# Example usage:
pre_average_calc_with_opponent(colnames)
```

## Previous Game Total Wins when matching against a specific opponent

For example: PHX vs BOS How many time has PHX win



```python
df.loc[:, ["Team", "Match Up", "Game Date", "W/L"]]
```

```python
# calculate the previous total wins
prev_total_wins = []

for team_name in df.Team.unique():
    subset = df[df['Team'] == team_name].reset_index(drop=True)
    for idx, row in subset.iterrows():
        if idx == 0:
            prev_total_wins.append(0)
        else:
            prev_row = subset.loc[:idx - 1]
            current_opponent = row['Opponent']
            prev_row = prev_row[prev_row['Opponent'] == current_opponent]
            prev_wins = np.sum(prev_row.loc[:, "W/L"])
            prev_total_wins.append(prev_wins)

df["Pre_Total_Wins"] = prev_total_wins
```

## Weights

Assign better weights to recent games

```python
# Weights

def assign_linear_weights(team, game_date, total_games):
    team_games = df[df['Team'] == team]
    team_games = team_games.sort_values(by='Game Date')

    team_games['Weight'] = range(1, len(team_games) + 1)

    return team_games


df['Weight'] = 0


for idx, row in df.iterrows():
    team = row['Team']
    game_date = row['Game Date']

    weighted_games = assign_linear_weights(team, game_date, len(df))

    weight = weighted_games.loc[weighted_games['Game Date'] == game_date, 'Weight'].values[0]
    df.at[idx, 'Weight'] = weight
```
